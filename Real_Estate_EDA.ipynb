 IQR Method (SalePrice)

```python
Q1 = df['SalePrice'].quantile(0.25)
Q3 = df['SalePrice'].quantile(0.75)

IQR = Q3 - Q1

outliers = df[
    (df['SalePrice'] < Q1 - 1.5 * IQR) |
    (df['SalePrice'] > Q3 + 1.5 * IQR)
]

outliers.shape
```

-Univariate Analysis (Single Variable) -SalePrice (Numerical)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))
counts, bins, patches = plt.hist(df['SalePrice'], bins=30)
colors = plt.cm.viridis(np.linspace(0, 1, len(patches)))
for patch, color in zip(patches, colors):
    patch.set_facecolor(color)
plt.title("Sale Price Distribution")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```


Insights:-

 -Sale prices are right-skewed, meaning most houses fall in a lower price range.

 -Few houses have very high prices, indicating premium or luxury properties.

 -Presence of extreme values suggests potential outliers.

 -House prices are concentrated in the mid-range, with few high-value outliers.

-GarageArea (Numerical)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))
counts, bins, patches = plt.hist(df['GarageArea'].dropna(), bins=30)
colors = plt.cm.plasma(np.linspace(0, 1, len(patches)))
for patch, color in zip(patches, colors):
    patch.set_facecolor(color)

plt.title("Distribution of Garage Area", fontsize=14)
plt.xlabel("Garage Area (sq ft)")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Insights:

  Most houses have small to medium-sized garages.

The distribution is right-skewed.

Very large garages are rare and appear as outliers.

Some houses have zero or very low garage area, indicating no garage.

Garage area is mostly moderate, with a few large-garage outliers.
